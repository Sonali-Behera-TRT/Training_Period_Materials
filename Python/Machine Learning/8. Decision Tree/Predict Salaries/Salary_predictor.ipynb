```python
import pandas as pd
```

```python
df = pd.read_csv('salaries.csv')
df.head()
```

```python
company_dummies = pd.get_dummies(df.company)
job_dummies = pd.get_dummies(df.job)
degree_dummies = pd.get_dummies(df.degree)
```

```python
merged = pd.concat([df, company_dummies, job_dummies, degree_dummies], axis = 'columns')
```

```python
merged
```

```python
final = merged.drop(['company', 'job', 'degree', 'abc pharma', 'business manager', 'masters'], axis = 'columns')
final
```

```python
x = final.drop('salary_more_then_100k', axis = 'columns')
y = final['salary_more_then_100k']
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=10)
```

```python
len(x_train)
```

```python
len(x_test)
```

```python
from sklearn import tree
model = tree.DecisionTreeClassifier()
model.fit(x_train, y_train)
```

```python
model.score(x_test, y_test)
```

```python
model.predict(x_test)
```

```python
y_test
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, model.predict(x_test))
cm
```

```python
model.predict_proba(x_test)
```
