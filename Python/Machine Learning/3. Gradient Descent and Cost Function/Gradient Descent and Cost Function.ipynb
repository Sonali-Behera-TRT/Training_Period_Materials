# Question 2

```python
import math

x = np.array([98, 68, 81, 80, 83, 52, 66, 30, 68, 73])
y = np.array([92, 56, 88, 70, 80, 49, 65, 35, 66, 67])
```

```python
def gradient_descent(x, y):
    m_curr = 0
    c_curr = 0
    learning_rate = 0.0005
    iteration = 1000
    n = len(x)
    prev_cost = 0
    for i in range(iteration):
        y_pred = m_curr * x + c_curr
        cost = (1 / n) * sum([j**2 for j in (y - y_pred)])
        md = -(2 / n) * sum(x * (y - y_pred))
        cd = -(2 / n) * sum(y - y_pred)
        m_curr = m_curr - learning_rate * md
        c_curr = c_curr - learning_rate * cd
        
        #print(f'prev: {prev_cost}, curr: {cost}, diff: {abs(prev_cost - cost)}')
        if(math.isclose(cost, prev_cost, rel_tol = 1e-20)):
            break
        prev_cost = cost
        print(f'm:{m_curr}, c:{c_curr}, cost:{cost}, iteration:{i}')
        
    return m_curr, c_curr
```

```python
m, c = gradient_descent(x, y)
print(m, c)
```
