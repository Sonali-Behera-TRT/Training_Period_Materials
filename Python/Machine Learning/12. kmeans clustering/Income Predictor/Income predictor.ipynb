```python
import pandas as pd
df = pd.read_csv('income.csv')
df
```

```python
import matplotlib.pyplot as plt
plt.scatter(df['Age'], df['Income($)'])
```

```python
from sklearn.cluster import KMeans
km = KMeans(n_clusters=3)
y_predicted = km.fit_predict(df[['Age', 'Income($)']])
y_predicted
```

```python
df['cluster'] = y_predicted
df
```

```python
df0 = df[df.cluster == 0]
df1 = df[df.cluster == 1]
df2 = df[df.cluster == 2]
```

```python
df2
```

```python
plt.scatter(df0['Age'], df0['Income($)'], color='red', label='cluster 1')
plt.scatter(df1['Age'], df1['Income($)'], color='blue', label='cluster 2')
plt.scatter(df2['Age'], df2['Income($)'], color='green', label='cluster 3')
plt.xlabel('Age')
plt.ylabel('Income($)')
plt.legend()
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
scaler.fit(df[['Income($)']])
df['Income($)'] = scaler.transform(df[['Income($)']])
```

```python
df
```

```python
scaler.fit(df[['Age']])
df['Age'] = scaler.transform(df[['Age']])
df
```

```python
km = KMeans(n_clusters=3)
y_predicted = km.fit_predict(df[['Age', 'Income($)']])
y_predicted
```

```python
df['cluster'] = y_predicted
```

```python
df
```

```python
centroids = km.cluster_centers_
```

```python
df0 = df[df.cluster == 0]
df1 = df[df.cluster == 1]
df2 = df[df.cluster == 2]
```

```python
df2
```

```python
plt.scatter(df0['Age'], df0['Income($)'], color='red', label='cluster 0')
plt.scatter(df1['Age'], df1['Income($)'], color='blue', label='cluster 1')
plt.scatter(df2['Age'], df2['Income($)'], color='green', label='cluster 2')
plt.scatter(centroids[:, 0], centroids[:, 1], color='yellow', label='centroids', marker='*')
plt.legend()
plt.xlabel('Age')
plt.ylabel('Income($)')
```

```python
sse = []
k_range = range(1, 10)

for i in k_range:
    km = KMeans(n_clusters=i)
    km.fit(df[['Age', 'Income($)']])
    sse.append(km.inertia_)
    
sse
```

```python
plt.plot(k_range, sse)
plt.xlabel('K')
plt.ylabel('SSE (Sum of Squared Error)')
```
