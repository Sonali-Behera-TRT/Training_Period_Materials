```python
import pandas as pd
```

```python
df = pd.read_csv('heart.csv')
df
```

```python
df.shape
```

```python
df.columns
```

```python
df[df.RestingBP > df.RestingBP.mean() + 3 * df.RestingBP.std()]
```

```python
df = df[df.RestingBP <= df.RestingBP.mean() + 3 * df.RestingBP.std()]
df.shape
```

```python
df[df.Cholesterol > df.Cholesterol.mean() + 3 * df.Cholesterol.std()]
```

```python
df = df[df.Cholesterol <= df.Cholesterol.mean() + 3 * df.Cholesterol.std()]
df.shape
```

```python
df[df.FastingBS > df.FastingBS.mean() + 3 * df.FastingBS.std()]
```

```python
df[df.MaxHR > df.MaxHR.mean() + 3 * df.MaxHR.std()]
```

```python
df[df.Oldpeak > df.Oldpeak.mean() + 3 * df.Oldpeak.std()]
```

```python
df = df[df.Oldpeak <= df.Oldpeak.mean() + 3 * df.Oldpeak.std()]
df.shape
```

```python
df[df.HeartDisease > df.HeartDisease.mean() + 3 * df.HeartDisease.std()]
```

```python
df.shape
```

```python
df = pd.get_dummies(df, drop_first=True)
df.head()
```

```python
df.shape
```

```python
x = df.drop('HeartDisease', axis = 'columns')
y = df.HeartDisease
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)
x_scaled
```

```python
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression

model_params = {
    'svm': {
        'model': SVC(),
        'params': {
            'kernel': ['rbf', 'linear', 'poly', 'sigmoid'],
            'C': [1, 10, 20],
            'gamma': ['auto', 'scale']
        }
    },
    'random_forest': {
        'model': RandomForestClassifier(),
        'params': {
            'n_estimators': [100, 200, 300],
            'criterion': ['gini', 'entropy', 'log_loss'],
            
        }
    },
    'logistric_regression': {
        'model': LogisticRegression(solver='liblinear', multi_class='auto'),
        'params': {
            'C': [1, 5, 10],
            'penalty': ['l1', 'l2', 'elasticnet', None],
            'dual': [True, False],
            'fit_intercept': [True, False]
        }
    }
}
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x_scaled, y)
```

```python
from sklearn.model_selection import GridSearchCV
scores = []

for model, mp in model_params.items():
    model = GridSearchCV(mp['model'], mp['params'], return_train_score=False)
    model.fit(x_train, y_train)
    scores.append({
        'model': model,
        'best_score': model.best_score_,
        'best_params': model.best_params_
    })
    
pd.DataFrame(scores)
```

```python
from sklearn.decomposition import PCA
pca = PCA(0.95)

x_pca = pca.fit_transform(x_scaled)
x_pca.shape
```
