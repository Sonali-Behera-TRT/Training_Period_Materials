```python
import pandas as pd
```

```python
df = pd.read_csv('./heights.csv')
df.head()
```

```python
df.sample(5)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.hist(df.height, rwidth= 0.8)
plt.xlabel('Height')
plt.ylabel('Count')
```

```python
min_limit, max_limit = df.height.quantile([0.05, 0.95])
```

```python
min_limit, max_limit
```

```python
df[(df.height < min_limit) | (df.height > max_limit)]
```

```python
df.shape
```

```python
df = df[(df.height >= min_limit) & (df.height <= max_limit)]
df.shape
```

```python
df.sample(10)
```
